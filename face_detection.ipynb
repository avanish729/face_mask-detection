```python
import os
```

```python
from keras.preprocessing import image
```

```python
import cv2
```

```python
categories=['with_mask','without_mask']
```

```python
data=[]
for category in categories:
    path=os.path.join('data1',category)
    for file in os.listdir(path):
        img_path=os.path.join(path,file)
        label=categories.index(category)
        img=cv2.imread(img_path)
        img=cv2.resize(img,(224,224))
        data.append([img,label])
        print(img.shape)
        # break
    # print (path)
```

```python
len(data)
```

```python
import random
```

```python
random.shuffle(data)
```

```python
X=[]
y=[]

for feature,label in data:
    X.append(feature)
    y.append(label)
    
```

```python
len(y)
```

```python
import numpy as np

X=np.array(X)
y=np.array(y)
```

```python
X.shape
```

```python
y.shape
```

```python
y
```

```python
X=X/255
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.2)
```

```python
X_test.shape
```

```python
X_train.shape
```

```python
from keras.applications.vgg16 import VGG16
```

```python
vgg=VGG16()
```

```python
vgg.summary()
```

```python
from keras import Sequential
```

```python
model =Sequential()
```

```python
for layer in vgg.layers[:-1] :
    model.add(layer)
    
```

```python
model.summary()
```

```python
#we freeze all layers so that they weights do not change while training our model
for layer in model.layers:
    layer.trainable=False
```

```python
model.summary()
```

```python
#now we add last layer in our model

from keras.layers import Dense
model.add(Dense(1,activation='sigmoid'))
```

```python
model.summary()
```

```python
model.compile(optimizer='Adam',loss='binary_crossentropy',metrics=['accuracy'])
```

```python
model.fit(X_train,y_train,epochs=5,validation_data=(X_test,y_test))
```

```python
import cv2 

# Load the pre-trained Haar cascade classifier for face detection
haar_cascade = cv2.CascadeClassifier(cv2.data.haarcascades + 'haarcascade_frontalface_default.xml')

cap = cv2.VideoCapture(0)

while True:
    ret, frame = cap.read()

    # Check if the frame is read successfully
    if not ret:
        print("Error: Unable to capture frame.")
        break

    # Resize the frame
    if frame is not None:  # Check if the frame is not empty
        frame = cv2.resize(frame, (224, 224))
        
        # Convert the frame to grayscale for face detection
        gray_frame = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)
        
        # Detect faces in the grayscale frame
        face_coordinates = haar_cascade.detectMultiScale(gray_frame)
        
        # Draw rectangles around the detected faces
        for (x, y, w, h) in face_coordinates:
            cv2.rectangle(frame, (x, y), (x+w, y+h), (0, 255, 0), 2)
        
        # Call the detection method for face mask detection
        y_pred = detect_face_mask(frame)
        
        # Draw label based on face mask detection result
        if y_pred == 0:
            draw_label(frame, "mask", (30, 30), (0, 255, 0))
        else:
            draw_label(frame, "no mask", (30, 30), (0, 0, 255))
        
        # Display the frame
        cv2.imshow("window", frame)

    if cv2.waitKey(1) & 0xFF == ord('x'):
        break

cap.release()
cv2.destroyAllWindows()
```
